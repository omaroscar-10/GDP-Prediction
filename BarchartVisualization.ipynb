```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter



data = pd.read_csv("data/output.csv")
df = pd.DataFrame(data)

# Filter countries where RequiredGDP is greater than GDP2022
df_filtered = df[(df['RequiredGDP'] > df['GDP2022']) & (df['flag70'] == 0)]

# Exclude Namibia from the plot
df_filtered_no_outlier = df_filtered[df_filtered['Country'] != 'Namibia']
#print(df_filtered_no_outlier.head(5))
fig, ax = plt.subplots(figsize=(10, 6))
bars_required = ax.barh(df_filtered_no_outlier['Country'], df_filtered_no_outlier['RequiredGDP'], label='Required GDP')
bars_current =ax.barh(df_filtered_no_outlier['Country'], df_filtered_no_outlier['GDP2022'], label='Current GDP', alpha=0.8)
plt.ylabel('Country')
plt.xlabel('GDP per capita (current US$)')
plt.title('Comparison of Required GDP for LE of 70 year and Current GDP (2022)')
#plt.text(20, 20, 'Countries with current LE less than 70 year', ha='center', va='center', fontsize=10, color='gray')
plt.text(10000,30,'Countries with current LE less than 70 year', ha='center', va='center', fontsize=10, color='red')

plt.legend()
required_gdp=list(df_filtered_no_outlier['RequiredGDP'])
# Function to format the tooltip text
x = range(len(required_gdp))
current_gdp=list(df_filtered_no_outlier['GDP2022'])
# Add values inside the bars

plt.tight_layout()
file_path =  'data/Plots/70year_Barchart.png'
plt.savefig(file_path)
plt.show()
```
